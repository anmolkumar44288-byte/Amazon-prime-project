**Import Libraries**

```python
import pandas as pd
import numpy as np
```

**Load Dataset**

```python
chicago = pd.read_csv("/content/Chicago.csv")
new_orleans = pd.read_csv("/content/New_Orleans.csv")
```

Add city column:

```python
chicago['city'] = 'Chicago'
new_orleans['city'] = 'New Orleans'
```

Merge dataset

```python
airbnb = pd.concat([chicago, new_orleans], ignore_index=True)
```

**Select Important Columns**

```python
columns = [
    'id','name','host_id','host_name',
    'neighbourhood_group',
    'neighbourhood',
    'latitude','longitude',
    'room_type',
    'price','minimum_nights',
    'number_of_reviews','last_review',
    'reviews_per_month',
    'availability_365',
    'city'
]

airbnb = airbnb[columns]
```

**Rename Columns**

```python
airbnb.rename(columns={
    'neighbourhood_group_cleansed':'neighbourhood_group',
    'neighbourhood_cleansed':'neighbourhood'
}, inplace=True)
```

**Clean Price Column**

```python
airbnb['price'] = (
    airbnb['price']
    .replace('[\$]', '', regex=True)
    .astype(float)
)
```

**Handle Missing Values**

Check null values:

```python
airbnb.isnull().sum()
```

Fill or remove missing values:

```python
airbnb['reviews_per_month'] = airbnb['reviews_per_month'].fillna(0)
```

**Remove rows with missing neighbourhoods:**

```python
airbnb.dropna(subset=['neighbourhood'], inplace=True)
```

**Remove Duplicates**

```python
airbnb.drop_duplicates(inplace=True)
```

**Standardize Text Columns**

```python
airbnb['neighbourhood'] = airbnb['neighbourhood'].str.title()
airbnb['room_type'] = airbnb['room_type'].str.title()
```

**Create Extra Columns**

```python
def price_category(x):
    if x < 100:
        return "Budget"
    elif x < 300:
        return "Mid-Range"
    else:
        return "Luxury"

airbnb['price_category'] = airbnb['price'].apply(price_category)
```

**Remove Outliers**

```python
q1 = airbnb['price'].quantile(0.25)
q3 = airbnb['price'].quantile(0.75)

iqr = q3 - q1

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

df = airbnb[(airbnb['price'] >= lower) & (airbnb['price'] <= upper)]
```

**Data Transformation**

Create Price Categories

```python
def price_category(x):
    if x < 100:
       return 'Budget'
    elif x < 300:
       return 'Medium'
    else:
       return 'Luxury'

airbnb['price_category'] = airbnb['price'].apply(price_category)
```

**Review Category**

```python
def review_category(x):
    if x < 10:
        return "Low Reviews"
    elif x < 50:
        return "Medium Reviews"
    else:
        return "High Reviews"

airbnb['review_category'] = airbnb['number_of_reviews'].apply(review_category)
```

**Host Type**


```python
host_counts = airbnb.groupby('host_id')['id'].transform('count')

airbnb['host_type'] = np.where(
    host_counts > 5,
    'Professional Host',
    'Individual Host'
)
```

**Export Cleaned Data**

```python
airbnb.to_csv("airbnb_cleaned.csv", index=False)
```
